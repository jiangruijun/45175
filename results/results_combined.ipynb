***Data retrieved from <span style="color:blue"> scenario 1 experiment 2 </span> as follows:***

```python
# Load the data
with open('s1e1_single_flow_cubic.json') as f1:
    data_s1e1_single_flow_cubic = json.load(f1)

df_s1e1_single_flow_cubic = pd.json_normalize(data_s1e1_single_flow_cubic['intervals'], 'streams')

# Round and convert 'start' columns to integers
df_s1e1_single_flow_cubic["end"] = df_s1e1_single_flow_cubic["end"].round().astype(int)

# Convert bits per second to megabits per second
df_s1e1_single_flow_cubic['throughput_mbps'] = df_s1e1_single_flow_cubic['bits_per_second'] / 1e6

# Convert rtt microseconds (µs) to rtt milliseconds (ms)
df_s1e1_single_flow_cubic['rtt_ms'] = df_s1e1_single_flow_cubic['rtt'] / 1000

# Convert cwnd (bytes) to cwnd (megabytes)
df_s1e1_single_flow_cubic['snd_cwnd_mb'] = df_s1e1_single_flow_cubic['snd_cwnd'] / 1e6

# Add the 'cc' column
df_s1e1_single_flow_cubic["cc"] = "CUBIC"

# Drop columns
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="start")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="seconds")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="bytes")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="rttvar")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="pmtu")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="omitted")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="sender")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="bits_per_second")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="rtt")
df_s1e1_single_flow_cubic = df_s1e1_single_flow_cubic.drop(columns="snd_cwnd")

df_s1e1_single_flow_cubic.head()
```

###### Jain’s Fairness Index:

$$ FI = \frac{\left( \sum_{i=1}^{n} T_i \right)^2}{n \sum_{i=1}^{n} T_i^2} $$


where:
* FI is the Fairness Index, with values between 0 and 1.
* 𝑛 is the total number of flows.
* 𝑇1, 𝑇2, . . . , 𝑇𝑛 are the measured throughput of individual flows.

```python
# Compute the Fairness Index for parallel flows
def fairnessIndexOfCompetingFlows(df):
    
    # Compute the mean throughput for each parallel flow
    mean_host_throughput_mbps = df.groupby('host')['throughput_mbps'].mean()
    
    # Convert mean throughput from Mbps to bps
    mean_host_throughput_bps = mean_host_throughput_mbps * 1e6
    
    # Number of flows
    n = len(mean_host_throughput_bps)
    
    # Compute the Fairness Index using Jain's formula
    fi = (mean_host_throughput_bps.sum())**2 / (n * (mean_host_throughput_bps**2).sum())
    
    return fi
```

```python
fi_competing_flows_cubic = fairnessIndexOfCompetingFlows(df_mean_host_throughput_cubic)
print(fi_competing_flows_cubic)
```

```python
fi_competing_flows_bbr = fairnessIndexOfCompetingFlows(df_mean_host_throughput_bbr)
print(fi_competing_flows_bbr)
```
